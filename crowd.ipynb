```python
from ultralytics import YOLO
from pathlib import Path
import pandas as pd

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ø§Ù„Ù…Ø¯Ø±Ø¨
model = YOLO("/kaggle/input/best-model-pt/best_Model.pt")  # ØªØ£ÙƒØ¯ÙŠ Ø¥Ù† Ø§Ø³Ù… Ø§Ù„Ù…Ù„Ù Ù†ÙØ³ Ø§Ù„Ù„ÙŠ Ø¹Ù†Ø¯Ùƒ

# ØªØ­Ø¯ÙŠØ¯ Ù…Ø¬Ù„Ø¯ Ø§Ù„ØµÙˆØ±
image_dir = Path("crowd-counting-dataset-1/test/images")

# Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø¹Ø¯Ø¯ Ø§Ù„Ø£Ø´Ø®Ø§Øµ Ù…Ù† ÙƒÙ„ ØµÙˆØ±Ø©
results_data = []

for image_file in sorted(image_dir.glob("*.jpg")):
    results = model(image_file)
    num_people = sum([1 for r in results[0].boxes.cls if int(r) == 0])  # class 0 = person
    results_data.append({
        "image": image_file.name,
        "person_count": num_people
    })

# ØªØ­ÙˆÙŠÙ„Ù‡Ø§ Ø¥Ù„Ù‰ DataFrame
df = pd.DataFrame(results_data)
df.head()
```

```python
def get_congestion_label(count):
    if count <= 10:
        return "Ø®ÙÙŠÙ"
    elif count <= 30:
        return "Ù…ØªÙˆØ³Ø·"
    else:
        return "Ø¹Ø§Ù„ÙŠ"

df["congestion_level"] = df["person_count"].apply(get_congestion_label)
df.head()
```

```python
import os
from shutil import copy

os.makedirs("cnn_dataset/Ø®ÙÙŠÙ", exist_ok=True)
os.makedirs("cnn_dataset/Ù…ØªÙˆØ³Ø·", exist_ok=True)
os.makedirs("cnn_dataset/Ø¹Ø§Ù„ÙŠ", exist_ok=True)

for _, row in df.iterrows():
    src = image_dir / row["image"]
    dst = Path(f"cnn_dataset/{row['congestion_level']}") / row["image"]
    copy(src, dst)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Ø¥Ø¹Ø¯Ø§Ø¯ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
datagen = ImageDataGenerator(validation_split=0.2, rescale=1./255)

train_data = datagen.flow_from_directory(
    "cnn_dataset",
    target_size=(128, 128),
    batch_size=16,
    class_mode="categorical",
    subset="training"
)

val_data = datagen.flow_from_directory(
    "cnn_dataset",
    target_size=(128, 128),
    batch_size=16,
    class_mode="categorical",
    subset="validation"
)

# Ù†Ù…ÙˆØ°Ø¬ CNN Ø¨Ø³ÙŠØ·
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax')  # 3 ÙØ¦Ø§Øª: Ø®ÙÙŠÙØŒ Ù…ØªÙˆØ³Ø·ØŒ Ø¹Ø§Ù„ÙŠ
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.fit(train_data, epochs=10, validation_data=val_data)
```

```python
model.fit(train_data, epochs=10, validation_data=val_data)
```

```python
import matplotlib.pyplot as plt


# Ù†Ø­ÙØ¸ Ù†ØªØ§Ø¦Ø¬ Ø§Ù„ØªØ¯Ø±ÙŠØ¨
history = model.fit(train_data, epochs=10, validation_data=val_data)

# Ù†Ø¹Ø±Ø¶ Ø¯Ù‚Ø© Ø§Ù„ØªØ¯Ø±ÙŠØ¨ ÙˆØ§Ù„ØªØ­Ù‚Ù‚
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('CNN Accuracy Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('CNN Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
